### Problem 1 :

By performing matrix multiplication between an \((n, k)\) matrix \(A\) and a \((k, m)\) matrix \(B\), let the resulting \((n, m)\) matrix be \(C\). The elements \(c_{ij}\) in matrix \(C\) are given by:

$$ c_{ij} = \sum_{l=1}^{k} a_{il} \cdot b_{lj} $$

### No. of Multiplications

For each element \(c_{ij}\) in matrix \(C\):
- **Number of Multiplications:** we need to compute \(k\) products.

Since there are (n * m) elements in the resultant matrix \(C\):
- **Total Multiplications:** Each element requires \(k\) multiplications.

  $$ \text{Total No of Multiplications} = n \times m \times k $$

### No. of Additions

For each element \(c_{ij}\) in matrix \(C\):
- **Number of Additions:** Adding \(k\) terms requires \(k - 1\) additions.

Since there are (n * m ) elements in matrix \(C\):
- **Total Additions:** Each element requires \(k - 1\) additions.

$$ \text{Total Additions} = n \times m \times (k - 1) $$

### Problem 2 :

```python
import time

def matrix_multiplication(A, B):
    n = len(A)
    k = len(A[0])
    m = len(B[0])


    C = [[0] * m for _ in range(n)]

    for i in range(n):
        for j in range(m):
            sum = 0
            for l in range(k):
                sum += A[i][l] * B[l][j]
            C[i][j] = sum

    return C


A = [
    [1,323,987,678,9,6],
    [4545,9345,65678,367,2777,4772],
    [723,65678,92345,451,877,314]
]
B = [
    [1,14,9,],
    [15,76,45],
    [67,89,43],
    [5,8,2],
    [8,7,4],
    [56,23,54]
]


start_time = time.time()
C = matrix_multiplication(A, B)
end_time = time.time()
print("A*B =")
print(C)
print("Time taken (Python):", end_time - start_time)

```

### Problem 4 :

Given function:  $$( f(x, y) = x^2 y + y^3 \sin(x) ).$$

The partial derivative of \( f(x, y) \) with respect to \( x \) is:

$$
\frac{\partial f}{\partial x} = 2xy + y^3 \cos(x)
$$

The partial derivative of ( f(x, y) )  with respect to ( y ) is:

$$
\frac{\partial f}{\partial y} = x^2 + 3y^2 \sin(x)
$$

The gradient of ( f(x, y) ) is:

$$
\nabla f = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = \left( 2xy + y^3 \cos(x), x^2 + 3y^2 \sin(x) \right)
$$

```python
import pandas as pd
```

```python
matrix = np.random.uniform(1, 2, size=(20, 5))
print(matrix)
data_frame = pd.DataFrame(matrix, columns=['a','b','c','d','e'])

std_devs = data_frame.std()
max_std_col = std_devs.idxmax()

row_means = data_frame.mean(axis=1)
min_mean_row = row_means.idxmin()

print("\nStandard Deviations for each column:")
print(std_devs)
print("\nMeans for each row:")
print(row_means)
print("\nColumn with the highest standard deviation:")
print(max_std_col)
print("\nRow with the lowest mean:")
print(min_mean_row)
```
